```python
#making binary classification for the response variable
#Dividing wine as good and bad by giving the limit for the quality
bins = (2, 6.5, 8)
group_names = ['bad', 'good']
wine['quality'] = pd.cut(wine['quality'], bins = bins, labels = group_names)
```

```python
#Now lets assign a labels to our quality variable
label_quality = LabelEncoder()
```

```python
#Bad becomes 0 and good becomes 1
wine['quality'] = label_quality.fit_transform(wine['quality'])
```

```python
wine['quality'].value_counts()
```

```python
sns.countplot(wine['quality'])
```

```python
sns.pairplot(wine,hue='quality')
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(wine.drop('quality',axis=1))
```

```python
scaled_features = scaler.fit_transform(wine.drop('quality',axis=1))
```

```python
df_feat = pd.DataFrame(scaled_features,columns=wine.columns[:-1])
```

```python
df_feat.head()
```

```python
X = df_feat
```

```python
y = wine['quality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
import tensorflow as tf
```

```python
df_feat.columns
```

```python
fixed_acidity = tf.feature_column.numeric_column("fixed_acidity")
volatile_acidity = tf.feature_column.numeric_column("volatile_acidity")
citric_adid = tf.feature_column.numeric_column("citric_acid")
residual = tf.feature_column.numeric_column('residual_sugar')
chlorides = tf.feature_column.numeric_column('chlorides')
free_sulfur_dioxide = tf.feature_column.numeric_column('free_sulfur_dioxide')
total_sulfur_dioxide = tf.feature_column.numeric_column('total_sulfur_dioxide')
density = tf.feature_column.numeric_column('density')
pH = tf.feature_column.numeric_column('pH')
sulphates = tf.feature_column.numeric_column('sulphates')
alcohal = tf.feature_column.numeric_column('alcohol')
```

```python
feat_cols = [fixed_acidity, volatile_acidity, citric_adid, residual, chlorides, free_sulfur_dioxide, total_sulfur_dioxide, density, pH, sulphates, alcohal]
```

```python
classifier = tf.estimator.DNNClassifier(hidden_units=[10,20,20,10], n_classes=2, feature_columns=feat_cols)
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train, y=y_train, batch_size=20, shuffle=True)
```

```python
classifier.train(input_fn=input_func, steps=500)
```

```python
#Model Evaluation 
pred_fn = tf.estimator.inputs.pandas_input_fn(x=X_test,batch_size=len(X_test), shuffle=False)
```

```python
note_predictions = list(classifier.predict(input_fn=pred_fn))
```

```python
note_predictions[0]
```

```python
final_preds = []
for pred in note_predictions:
    final_preds.append(pred['class_ids'][0])
```

```python
print(confusion_matrix(y_test,final_preds))
```

```python
print(classification_report(y_test,final_preds))
```

```python
#perform machine learnung Algorithm
#Random Forest Classifier
rfc = RandomForestClassifier(n_estimators=300)
rfc.fit(X_train,y_train)
```

```python
pred_rfc = rfc.predict(X_test)
```

```python
print(classification_report(y_test, pred_rfc))
```

Random Forest gives the accuracy of 89%


```python
print(confusion_matrix(y_test, pred_rfc))
```

```python
#Stochastic Gradient Descent Classifier: Is simple yet very efficient approach to fitting linear classifiers and regressor under convex loss functions
#such as linear support vector machines and Logistic Regression
sgd = SGDClassifier(penalty=None)
sgd.fit(X_train, y_train)
pred_sgd = sgd.predict(X_test)
```

```python
print(classification_report(y_test, pred_sgd))
```

85% accuracy using stochastic gradient descent Classifier

```python
print(confusion_matrix(y_test, pred_sgd))
```

```python
#Support Vector Classifier
#SVM are set of supervised machine learning methods used for classification,
#regression and outliers detections
#SVC,NuSVC and LinearSVC are capable of preforming binary and multiclass classifiction on a dataset
svc = SVC()
svc.fit(X_train, y_train)
pred_svc = svc.predict(X_test)
```

```python
print(classification_report(y_test, pred_svc))
```

Support Vector classifier gets 86% accuracy

```python
#Lets try to increase our accuracy of models using Grid Search CV
#grid search CV implements a "fit", "score", "predict" and "predict-proba" methods
# finding best parameters for our SVC model
param = {
    'C': [0.1,0.8,0.9,1,1.1,1.2,1.3,1.4],
    'kernel':['linear','rbf'],
    'gamma':[0.1,0.8,0.9,1.1,1.2,1.3,1.4]
}
grid_svc = GridSearchCV(svc, param_grid=param, scoring='accuracy', cv=10)
```

```python
grid_svc.fit(X_train, y_train)
```

```python
grid_svc.best_params_
```

```python
#Lets run our SVC again with the best parameters
svc2 = SVC(C = 1.3, gamma = 0.8, kernel = 'rbf')
svc2.fit(X_train, y_train)
pred_svc2 = svc2.predict(X_test)
print(classification_report(y_test, pred_svc2))
```

```python
#cross Validation Score for random forest and SGD
rfc_eval = cross_val_score(estimator=rfc,X=X_train, y=y_train,cv=10)
rfc_eval.mean()
```
